```python
import json
import requests
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
params = {
    "accept": "application/json", 
    "X-API-Token": os.environ['API_TOKEN'],
}
```

```python
symbol = 'BTC-USD'
```

```python
url = f"https://api.blockchain.com/v3/exchange/l2/{symbol}"
```

```python
data = requests.get(url, params=params).json()
```
